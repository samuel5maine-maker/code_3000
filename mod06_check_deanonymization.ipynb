### Describe your reaction to the match rate. What sources of data might be prone to privacy issues?

My match rate was 0.38, a shockingly high number given the barriers in place. This 0.38 is only unique one to one matches. If we instead extended this to, say, a gropu of potential canidates for each anon_id, we could get a match rate of 1.0 (I tested this). Meaning for every row in the anon ID there is at least one row in the auxillary list that has matching credentials, some just have multiple. It's a little scary thinking about the implications of how much you could narrow down any row in an anonymized data set to just a couple people.

Any data, or combination of data, that can be cross referenced with public records & serve as a primary key seems like it would be extremely effective at finding one to one matches. Locational and Birth Information I feel are especially prone, as if you have someones address you can easily look up there name, where they work, etc. So things like: Address info (even partial), Birthday/year, or Even Marriage status, could eventually lead to security issues

### Provide examples of at least two other variables that could increase the match rate if included.

Lots of things could be added: Birth Month, Marriage Status, Income Level, Presence of Criminal History (T or F), or even something as small as just turning zip3 into zip4 could go a long way increasing the match rate. The biggest problem with the match rate currently is that for a lot a rows the features could be applied to multiple people. So by adding more features or a feature with more possible options, this should become less and less of an issue.
